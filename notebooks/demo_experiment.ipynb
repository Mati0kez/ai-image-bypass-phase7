```python
import sys
sys.path.insert(0, 'src')
from benchmark.runner import BenchmarkRunner, BenchmarkConfig
from benchmark.report import generate_experiment_report
import tempfile
from pathlib import Path
from PIL import Image
import numpy as np

print('环境准备完成')
```

```python
# 创建临时测试图像
with tempfile.TemporaryDirectory() as tmp:
    input_dir = Path(tmp) / 'input'
    output_dir = Path(tmp) / 'output'
    input_dir.mkdir()
    for i in range(5):
        img = Image.fromarray(np.random.randint(0, 255, (64, 64, 3), dtype=np.uint8))
        img.save(input_dir / f'test_{i}.jpg')

    config = BenchmarkConfig(
        input_dir=str(input_dir),
        output_dir=str(output_dir),
        mode='experiment',
        platforms=['remote:mock'],
        sample_size=5,
        detector_feedback=False,
    )
    runner = BenchmarkRunner(config)
    results = runner.run()
    paths = generate_experiment_report(results, runner.failure_cases, str(output_dir), config.platforms)
    print('Experiment 完成！')
    print('JSON:', paths['json'])
    print('CSV:', paths['csv'])
    print('MD:', paths['md'])
```
